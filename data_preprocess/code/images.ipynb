```python
import numpy as np
import os
import pandas as pd
import cv2
from PIL import Image
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
```

```python
# Load VGG16 model pre-trained on ImageNet data
base_model = VGG16(weights='imagenet')
model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc1').output)

def extract_features(img_path, model):
    # print('Extracting features from image: ', img_path)
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    features = model.predict(img_array)
    return features
```

```python
raw_data = pd.read_csv('../../raw_data/train.csv')
```

```python
# Directory containing images
image_folder = '../../downloaded_images/'

# Extract features and also raw images
features = []
file_names = []
image_size = (224, 224)
images = []
for i in range(len(raw_data)):
    if os.path.isfile('../../downloaded_images/' + str(i) + '.jpg'):
        file_path = os.path.join(image_folder, str(i) + '.jpg')
        img_features = extract_features(file_path, model)
        features.append(img_features.flatten())

        figure = cv2.imread('../../downloaded_images/' + str(i) + '.jpg')
        figure = cv2.resize(figure, image_size)
        images.append(figure)
        file_names.append(i)
    elif os.path.isfile('../../downloaded_images/' + str(i) + '.png'):
        file_path = os.path.join(image_folder, str(i) + '.png')
        img_features = extract_features(file_path, model)
        features.append(img_features.flatten())

        figure = cv2.imread('../../downloaded_images/' + str(i) + '.png')
        figure = cv2.resize(figure, image_size)
        images.append(figure)
        file_names.append(i)
    elif os.path.isfile('../../downloaded_images/' + str(i) + '.jpeg'):
        file_path = os.path.join(image_folder, str(i) + '.jpeg')
        img_features = extract_features(file_path, model)
        features.append(img_features.flatten())

        figure = cv2.imread('../../downloaded_images/' + str(i) + '.jpeg')
        figure = cv2.resize(figure, image_size)
        images.append(figure)
        file_names.append(i)
    else:
        continue
    
features = np.array(features)
```

```python
# save the features
np.save('../df/images/images_features.npy', features)
np.save('../df/images/images.npy', images)
np.save('../df/images/file_names.npy', file_names)
```
